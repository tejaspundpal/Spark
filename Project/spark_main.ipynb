```python
def setup_logger(filename):
    if not os.path.exists(filename):
        with open(filename, 'w') as f:
            pass

    logger = logging.getLogger("sparkLogger")
    logger.setLevel(logging.INFO)

    if logger.hasHandlers():
        logger.handlers.clear()

    file_handler = logging.FileHandler(filename,mode="a")
    console_handler = logging.StreamHandler()
    formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
    file_handler.setFormatter(formatter)
    console_handler.setFormatter(formatter)
    logger.addHandler(file_handler)
    logger.addHandler(console_handler)
    logger.info("Logger setup completed")
    return logger

logger = setup_logger("spark.log")
```

```python
#Top 5 Sellers by Revenue
logger.info("Finding, displaying, and saving top 5 sellers by revenue")
top_5_sellers_df = total_sales_revenue_df.join(order_items_df,'order_id').join(sellers_df,'seller_id').groupBy('seller_id').agg(round(sum('Total price'),2).alias('Total_price'),round(sum('Total freight value'),2).alias('Total_freight_value')).orderBy(col('Total_price').desc()).limit(5)
top_5_sellers_df.display()
top_5_sellers_df.write.mode('overwrite').saveAsTable('top_5_sellers')
```

```python
#Top 3 products per category by revenue
from pyspark.sql.functions import rank
from pyspark.sql.window import Window

logger.info("Finding, displaying, and saving top 3 products per category by revenue using window function")
window_spec = Window.partitionBy('product_category_name').orderBy(col('Total_price').desc())
top_3_products_per_category_df = total_sales_revenue_df.withColumn('Total price and freight value',col('Total price')+col('Total freight value'))\
    .join(order_items_df,'order_id').join(products_df,'product_id').join(product_category_name_translation_df,'product_category_name')\
    .groupBy('product_category_name', 'product_id')\
    .agg(round(sum('Total price and freight value'),2).alias('Total_price'))\
    .withColumn('rank', rank().over(window_spec))\
    .filter(col('rank') <= 3)\
    .select('product_category_name','product_id','Total_price','rank')
top_3_products_per_category_df.display()
top_3_products_per_category_df.write.mode('overwrite').saveAsTable('top_3_products_per_category_with_rank')
```

```python
# fastest vs slowest order delivery times per seller.

logger.info("Finding, displaying, and saving fastest vs slowest order delivery times per seller")
delivery_time_df = orders_df.join(order_items_df,'order_id').join(sellers_df,'seller_id')\
    .groupBy('seller_id')\
    .agg(round(avg(col('order_delivered_customer_date').cast('double') - col('order_purchase_timestamp').cast('double')),2))\
    .withColumnRenamed('round(avg((CAST(order_delivered_customer_date AS DOUBLE) - CAST(order_purchase_timestamp AS DOUBLE))), 2)','avg_delivery_time')\
    .orderBy(col('avg_delivery_time').desc())
delivery_time_df.limit(5).display()
print(f'fastest delivery time per seller : {delivery_time_df.first()}')
print(f'slowest delivery time per seller : {delivery_time_df.orderBy(col("avg_delivery_time").asc()).first()}')
```

```python
#Customer Lifetime Value (CLV) = total revenue per customer.
logger.info("Finding, displaying, and saving Customer Lifetime Value (CLV) = total revenue per customer.")
customer_lifetime_value_df = total_sales_revenue_df.withColumn('Total price and freight value',col('Total price')+col('Total freight value'))\
    .join(orders_df,'order_id').join(customers_df,'customer_id')\
    .groupBy('customer_unique_id')\
    .agg(round(sum('Total price and freight value'),2).alias('CLV'))
customer_lifetime_value_df.display()
```

```python
# Monthly Active Customers (MAC) = distinct customers per month.
logger.info("Finding, displaying, and saving Monthly Active Customers (MAC) = distinct customers per month.")
monthly_active_customers_df = orders_df.groupBy(month('order_purchase_timestamp').alias('month'),year('order_purchase_timestamp').alias('year'))\
    .agg(countDistinct('customer_id').alias('MAC'))
monthly_active_customers_df.display()
```

```python
# churn customers → customers who ordered only once.
logger.info("Finding, displaying, and saving churn customers → customers who ordered only once.")
churn_customers_df = orders_df.groupBy('customer_id').agg(countDistinct('order_id').alias('orders')).filter(col('orders')==1)
churn_customers_df.display()
print(f'Number of churn customers : {churn_customers_df.count()}')
```

```python
# top 3 most reviewed products.
# Average rating per category.
# Distribution of ratings (1–5 stars).
# Identify products with most negative reviews.

#1
logger.info("Finding, displaying, and saving top 3 most reviewed products.")
top_3_most_reviewed_products_df = order_reviews_df.join(order_items_df,'order_id').join(products_df,'product_id')\
    .groupBy('product_id','product_category_name')\
    .agg(countDistinct('review_id').alias('review_count'))\
    .orderBy(col('review_count').desc())\
    .limit(3)
top_3_most_reviewed_products_df.display()

#2
logger.info("Finding, displaying, and saving average rating per category.")
avg_rating_per_category_df = order_reviews_df.join(order_items_df,'order_id').join(products_df,'product_id')\
    .groupBy('product_id','product_category_name')\
    .agg(avg('review_score').alias('avg_rating'))\
    .orderBy(col('avg_rating').desc())
avg_rating_per_category_df.display()

#3
logger.info("Finding, displaying, and saving distribution of ratings (1–5 stars).")
distribution_of_ratings_df = order_reviews_df.groupBy('review_score').agg(count('review_id').alias('count'))
distribution_of_ratings_df.display()

#4
logger.info("Finding, displaying, and saving identify products with most negative reviews.")
most_negative_reviews_df = order_reviews_df.join(order_items_df,'order_id').join(products_df,'product_id')\
    .groupBy('product_id','product_category_name')\
    .agg(count('review_id').alias('review_count'))\
    .orderBy(col('review_count').asc())\
    .limit(3)
most_negative_reviews_df.display()
```

```python
#Top 5 cities by order count
logger.info("Finding, displaying, and saving top 5 cities by order count.")
top_5_cities_by_order_count_df = orders_df.join(customers_df,'customer_id').groupBy('customer_city').agg(count('order_id').alias('order_count'))\
    .orderBy(col('order_count').desc()).limit(5)
top_5_cities_by_order_count_df.display()
#Top 5 cities by revenue
logger.info("Finding, displaying, and saving top 5 cities by revenue.")
top_5_cities_by_revenue_df = total_sales_revenue_df.withColumn('Total price and freight value',col('Total price')+col('Total freight value')).join(orders_df,'order_id').join(customers_df,'customer_id').join(order_items_df,'order_id')\
    .groupBy('customer_city')\
    .agg(sum('Total price and freight value').alias('revenue'))\
    .orderBy(col('revenue').desc()).limit(5)
top_5_cities_by_revenue_df.display()
```
